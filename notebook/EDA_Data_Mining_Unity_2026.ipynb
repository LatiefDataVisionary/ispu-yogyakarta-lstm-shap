```python
import pandas as pd
import numpy as np
```

```python
ispu_data_by_year = {}
base_url = 'https://raw.githubusercontent.com/LatiefDataVisionary/ispu-yogyakarta-lstm-shap/refs/heads/main/datasets/raw/ispu_'

for year in range(2020, 2025):
    file_name = f'ispu_{year}.csv'
    full_url = f'{base_url}{year}.csv'
    try:
        df = pd.read_csv(full_url)
        ispu_data_by_year[year] = df
        print(f'Successfully loaded data for year {year}')
    except Exception as e:
        print(f'Error loading data for year {year}: {e}')

# You can access individual dataframes like this:
# ispu_2020_df = ispu_data_by_year.get(2020)
# ispu_2021_df = ispu_data_by_year.get(2021)
```

```python
for year, df in ispu_data_by_year.items():
    print(f"\n{'='*20} Data Inspection for Year {year} {'='*20}")

    print("\n1. First 5 Rows:")
    display(df.head())

    print("\n2. Info:")
    df.info()

    print("\n3. Statistical Summary:")
    display(df.describe())

    print("\n4. Missing Values:")
    print(df.isnull().sum())

    print("\n5. Duplicate Rows:")
    print(f"Total duplicates: {df.duplicated().sum()}")
```

```python
# Concatenate all DataFrames into a single DataFrame
all_ispu_data = pd.concat(ispu_data_by_year.values(), ignore_index=True)

# Define a mapping for Indonesian month names to English month names
month_mapping = {
    'Januari': 'January',
    'Februari': 'February',
    'Maret': 'March',
    'April': 'April',
    'Mei': 'May',
    'Juni': 'June',
    'Juli': 'July',
    'Agustus': 'August',
    'September': 'September',
    'Oktober': 'October',
    'November': 'November',
    'Desember': 'December'
}

# Function to replace Indonesian month names with English ones
def replace_indonesian_months(date_str):
    if pd.isna(date_str): # Handle potential NaN values
        return date_str
    for id_month, en_month in month_mapping.items():
        date_str = date_str.replace(id_month, en_month)
    return date_str

# Apply the replacement function to the 'Waktu' column
all_ispu_data['Waktu'] = all_ispu_data['Waktu'].apply(replace_indonesian_months)

# Convert 'Waktu' column to datetime objects using a consistent format
all_ispu_data['Waktu'] = pd.to_datetime(all_ispu_data['Waktu'], format='%d %B %Y', errors='coerce')

# Display basic information about the merged DataFrame
print("\nMerged Data Info:")
all_ispu_data.info()

print("\nFirst 5 rows of the merged data:")
display(all_ispu_data.head())

print("\nLast 5 rows of the merged data:")
display(all_ispu_data.tail())
```

### 🚀 Comprehensive ISPU Data EDA - Stage 1: Feature Engineering & Cleaning
In this step, we will enrich the dataset by extracting temporal features (Month, Year, Day of Week) and handling any remaining missing values in the newer columns (PM2.5, HC) to ensure a smooth visualization process.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

# Set Plotting Style
sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (15, 8)

# Feature Engineering: Time-based features
all_ispu_data['Year'] = all_ispu_data['Waktu'].dt.year
all_ispu_data['Month'] = all_ispu_data['Waktu'].dt.month_name()
all_ispu_data['Day'] = all_ispu_data['Waktu'].dt.day_name()
all_ispu_data['Month_Num'] = all_ispu_data['Waktu'].dt.month

# Handling Missing Values for Visualization (Filling with 0 or Mean for trend analysis)
# We keep a copy for analysis where we don't want to distort statistical distributions
eda_df = all_ispu_data.copy()
eda_df[['PM2.5', 'HC']] = eda_df[['PM2.5', 'HC']].fillna(0)

print("Feature Engineering Complete.")
display(eda_df.head())
```

### 🌟 Stage 5: Air Quality Categorization (Standard ISPU)
In Indonesia, ISPU values are categorized into levels. Let's create a categorization logic based on the maximum value across all pollutants to determine the 'Air Quality Status' for every single day.

```python
def categorize_ispu(value):
    if value <= 50: return 'Baik (Good)'
    elif value <= 100: return 'Sedang (Moderate)'
    elif value <= 200: return 'Tidak Sehat (Unhealthy)'
    elif value <= 300: return 'Sangat Tidak Sehat (Very Unhealthy)'
    else: return 'Berbahaya (Hazardous)'

# Identify the maximum ISPU value per row (ignoring temporal/categorical columns)
eda_df['Max_ISPU'] = eda_df[numeric_cols].max(axis=1)
eda_df['AQI_Category'] = eda_df['Max_ISPU'].apply(categorize_ispu)

# Visualize the proportion of Air Quality Categories over the years
plt.figure(figsize=(14, 7))
sns.countplot(data=eda_df, x='Year', hue='AQI_Category', palette='viridis')
plt.title('Air Quality Category Frequency by Year', fontsize=16)
plt.ylabel('Number of Days')
plt.legend(title='Category', loc='upper right')
plt.show()
```

### 🌀 Stage 8: Advanced Time-Series Decomposition
Kita akan memecah sinyal polutan (PM10) menjadi tiga komponen utama: **Trend** (arah jangka panjang), **Seasonality** (pola berulang), dan **Residual** (gangguan/noise). Ini sangat penting untuk memahami apakah kualitas udara membaik atau memburuk secara fundamental selama 5 tahun terakhir.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Menyiapkan data deret waktu yang kontinu
# Memastikan hanya kolom numerik yang di-resample untuk menghindari TypeError
ts_df = eda_df.set_index('Waktu')[numeric_cols].resample('D').mean().interpolate()

# Melakukan dekomposisi pada PM10
result = seasonal_decompose(ts_df['PM10'], model='additive', period=365)

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(15, 12), sharex=True)
result.observed.plot(ax=ax1, color='blue', title='Observed PM10')
result.trend.plot(ax=ax2, color='red', title='Trend (Long-term Movement)')
result.seasonal.plot(ax=ax3, color='green', title='Seasonality (Annual Cycle)')
result.resid.plot(ax=ax4, color='gray', title='Residuals (Unexplained Noise)', style='.')
plt.tight_layout()
plt.show()
```

### 🛑 Stage 11: Outlier & Anomaly Detection
Kita akan menggunakan metode IQR untuk mendeteksi 'Extreme Pollution Events'. Kejadian ini biasanya berkaitan dengan peristiwa khusus (seperti kebakaran hutan, erupsi gunung merapi, atau lonjakan aktivitas kendaraan saat liburan).

```python
Q1 = eda_df['PM10'].quantile(0.25)
Q3 = eda_df['PM10'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

anomalies = eda_df[eda_df['PM10'] > upper_bound]

plt.figure(figsize=(15, 6))
sns.scatterplot(data=eda_df, x='Waktu', y='PM10', alpha=0.3, label='Normal Days')
sns.scatterplot(data=anomalies, x='Waktu', y='PM10', color='red', label='Anomaly/Extreme Events')
plt.axhline(upper_bound, color='orange', linestyle='--', label='Outlier Threshold')
plt.title(f'Detection of Extreme Air Pollution Events (Total: {len(anomalies)} days)', fontsize=16)
plt.legend()
plt.show()
```

### 📈 Stage 14: Time-Series Stationarity Test (Augmented Dickey-Fuller)
Sebelum masuk ke LSTM atau model forecasting lainnya, kita harus tahu apakah data ini 'Stasioner' (rata-rata dan varians stabil). Ini adalah langkah 'genius' yang sering dilewatkan dalam EDA biasa.

```python
from statsmodels.tsa.stattools import adfuller

def check_stationarity(timeseries):
    print('--- Results of Augmented Dickey-Fuller Test ---')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
    if dfoutput['p-value'] < 0.05:
        print('\nKesimpulan: Data STASIONER (Siap untuk modeling forecasting)')
    else:
        print('\nKesimpulan: Data TIDAK STASIONER (Perlu differencing/transformasi sebelum modeling)')

check_stationarity(ts_df_clean['PM10'])
```

### 🕰️ Stage 15: Extreme Event Persistence (Consecutive High Pollution Days)
Seberapa lama polusi buruk biasanya bertahan di Yogyakarta? Kita akan menghitung durasi hari-hari beruntun di mana polusi berada di atas rata-rata.

```python
threshold = eda_df['PM10'].mean()
eda_df['Is_High'] = eda_df['PM10'] > threshold

# Menghitung durasi beruntun
eda_df['Group'] = (eda_df['Is_High'] != eda_df['Is_High'].shift()).cumsum()
consecutive_days = eda_df[eda_df['Is_High']].groupby('Group').size()

plt.figure(figsize=(12, 6))
sns.histplot(consecutive_days, bins=20, kde=True, color='purple')
plt.title('Distribution of Consecutive High Pollution Days (Persistence Analysis)', fontsize=16)
plt.xlabel('Number of Consecutive Days above Mean')
plt.ylabel('Frequency')
plt.show()
```

### 🕵️ Stage 16: Cross-Correlation Analysis (Lead-Lag Effect)
Apakah lonjakan satu polutan memicu polutan lainnya beberapa hari kemudian? Kita akan menganalisis hubungan sebab-akibat temporal menggunakan *Cross-Correlation Function* (CCF).

```python
from statsmodels.tsa.stattools import ccf

# Menghitung cross-correlation antara NO2 dan O3 (0 sampai 30 hari)
cross_corr = ccf(ts_df_clean['NO2'], ts_df_clean['O3'], adjusted=False)[:31]

plt.figure(figsize=(12, 5))
plt.stem(range(len(cross_corr)), cross_corr)
plt.axhline(0.05, color='red', linestyle='--')
plt.axhline(-0.05, color='red', linestyle='--')
plt.title('Cross-Correlation: NO2 Leading O3 (0 to 30 Days Lag)', fontsize=16)
plt.xlabel('Lag (Days)')
plt.ylabel('Correlation Coefficient')
plt.show()
```

### 📻 Stage 17: Spectral Analysis (Fourier Transform)
Kita akan menggunakan algoritma FFT untuk menemukan siklus polusi tersembunyi yang mungkin tidak terlihat di domain waktu, seperti siklus mingguan yang sangat halus.

```python
from scipy.fft import fft, fftfreq

# Menghitung FFT pada PM10
n = len(ts_df_clean['PM10'])
y = ts_df_clean['PM10'].values - ts_df_clean['PM10'].mean()
yf = fft(y)
xf = fftfreq(n, 1)[:n//2]

plt.figure(figsize=(15, 6))
plt.plot(xf[1:], 2.0/n * np.abs(yf[1:n//2]))
plt.title('Power Spectral Density (Fourier Transform of PM10)', fontsize=16)
plt.xlabel('Frequency (Cycles per Day)')
plt.ylabel('Amplitude')
plt.axvline(1/7, color='green', linestyle=':', label='Weekly Cycle (1/7)')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.stattools import ccf

# Menghitung cross-correlation antara NO2 dan O3 (0 sampai 30 hari)
cross_corr = ccf(ts_df_clean['NO2'], ts_df_clean['O3'], adjusted=False)[:31]

plt.figure(figsize=(12, 5))
plt.stem(range(len(cross_corr)), cross_corr)
plt.axhline(0.05, color='red', linestyle='--')
plt.axhline(-0.05, color='red', linestyle='--')
plt.title('Cross-Correlation: NO2 Leading O3 (0 to 30 Days Lag)', fontsize=16)
plt.xlabel('Lag (Days)')
plt.ylabel('Correlation Coefficient')
plt.show()
```

### 📻 Stage 17: Spectral Analysis (Fourier Transform)
Kita akan mendeteksi siklus periodik yang tersembunyi dengan mengubah data ke dalam domain frekuensi.

```python
from scipy.fft import fft, fftfreq

# Menghitung FFT pada PM10
n = len(ts_df_clean['PM10'])
y = ts_df_clean['PM10'].values - ts_df_clean['PM10'].mean()
yf = fft(y)
xf = fftfreq(n, 1)[:n//2]

plt.figure(figsize=(15, 6))
plt.plot(xf[1:], 2.0/n * np.abs(yf[1:n//2]))
plt.title('Power Spectral Density (Fourier Transform of PM10)', fontsize=16)
plt.xlabel('Frequency (Cycles per Day)')
plt.ylabel('Amplitude')
plt.axvline(1/7, color='green', linestyle=':', label='Weekly Cycle (1/7)')
plt.legend()
plt.show()
```
